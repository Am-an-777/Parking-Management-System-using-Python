```python
import datetime

Max_Spaces = 10
vehicle_type = []
vehicle_number = []
entry_time = []
space = Max_Spaces

def validate_vehicle_number(vehicle_number):
    if len(vehicle_number)!= 10:
        return False
    if not vehicle_number[:2].isalpha() or not vehicle_number[2:4].isdigit() \
        or not vehicle_number[4:6].isalpha() or not vehicle_number[6:].isdigit():
        return False
    return True

while True:
    print("-------------------")
    print("SAFETY ZONE PARKING")
    print("-------------------")
    choice = input("Enter-1 for Vehicle Entry\nEnter-2 for Parked Vehicle Details\nEnter-3 for Vehicle Exit\nEnter-4 for Closing Service\n")

    if choice == '1':
        if space == 0:
            print("Sorry, the parking lot is full. Please try again later.\n")
            continue
        
        typee = True
        while typee:
            print("Vehicle Parking Prices/Hour: Bike=10 Rs/Hr\tCar=20 Rs/Hr\n")
            vtype = input("Enter-1 for Bike Entry\nEnter-2 for Car Entry\n")
            if vtype == '1':
                vehicle_type.append("Bike")
                space -= 1
                typee = False
            elif vtype == '2':
                vehicle_type.append("Car")
                space -= 1
                typee = False
            elif vtype.isalpha():
                print("Please Enter Valid Option\n")
            else:
                print("Please Enter Valid Option\n")

        no = True
        while no:
            vno = input("Enter Vehicle Number :\n").upper()
            if vno in vehicle_number:
                print("Vehicle Number Already Exists\n")
            elif validate_vehicle_number(vno):
                no = False
                vehicle_number.append(vno)
            else:
                print("Enter Vehicle no. in proper format\n")

        dt = True
        while dt:
            dt = False
            x = datetime.datetime.now()
            entry_time.append(x)
        print("Record details Saved\n")

    elif choice == '2':
        print("Parked Vehicle")
        print("---------------------------------------------------------------------------------------")
        print("Vehicle Number\t\tVehicle Type\t\tEntry-Time")
        print("---------------------------------------------------------------------------------------")
        for i in range(len(vehicle_number)):
            print(vehicle_number[i],"\t\t",vehicle_type[i],"\t\t",entry_time[i])
        print("---------------------------------------------------------------------------------------")
        print("Spaces Available:",space,"\n")

    elif choice == '3':
        if vehicle_number == []:
            print("Parking lot empty\n")
            continue
            
        no = True
        while no:
            vno = input("Enter Vehicle Number To Delete :\n").upper()
            if len(vno) == 10:
                if vno in vehicle_number:
                    i = vehicle_number.index(vno)
                    print("Removed Vehicle Details")
                    print("------------------------------------------------")
                    print("Vehicle Number:",vehicle_number.pop(i))
                    bc = vehicle_type.pop(i)
                    print("Vehicle Type:",bc)
                    space += 1
                    entry = entry_time.pop(i)
                    print("Entry Time:",entry)
                    exit = datetime.datetime.now()
                    print("Exit Time:",exit)

                    print("------------------------------------------------\n")
                    difference = exit - entry
                    f = difference.total_seconds()
                    hr = f // 3600
                    f = f % 3600
                    mi = f // 60
                    f = f % 60
                    sec = f
                    print("Total Hour(s):",int(hr),"Minute(s):",int(mi),"Second(s):",int(sec))
                    if int(hr) == 0 and bc == "Bike":
                        amt = 10
                        pay=not True
                    elif int(hr)==0 and bc=="Car":
                        amt=20
                        pay=not True
                    elif int(hr)>=1 and bc=="Bike":
                        amt=int(hr)*10
                        if int(mi)>=1 and bc=="Bike":
                            amt+=10
                            pay=not True
                    elif int(hr)>=1 and bc=="Bike":
                        amt=int(hr)*10
                        if int(sec)>=1 and bc=="Bike":
                            amt+=10
                            pay=not True
                    elif int(hr)>=1 and bc=="Car":
                        amt=int(hr)*20
                        if int(mi)>=1 and bc=="Car":
                            amt+=20
                            pay=not True
                    elif int(hr)>=1 and bc=="Car":
                        amt=int(hr)*20
                        if int(sec)>=1 and bc=="Car":
                            amt+=20
                            pay=not True
                    print("Total Charge:",amt,"Rs.\n")
                    print("Removed Successfully")
                    print("Spaces available after vehicle removal:",space)
                    break
                elif vno not in vehicle_number:
                    print("No Entry Available\n")
            else:
                print("Enter Valid Vehicle Number\n")
      
    elif choice=='4':
        print("Thank you for using our service\n")
        break
                
    elif choice.isalpha():
        print("Please Enter Valid Choice\n")
                
    else:
        print("Please Enter Valid Choice\n")                 
```
